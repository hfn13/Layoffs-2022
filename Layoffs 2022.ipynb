```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

layoffs = pd.read_csv('layoffs.csv')
layoffs
```

```python
layoffs.info()
```

```python
print(f'Total number of companies in the dataset are {len(layoffs.company.unique())}')
print()
print('The companies included are: ')
list(layoffs.company.unique())
```

```python
print(f'Total number of countries in dataset {len(layoffs.country.unique())}')
print()
print('Number of companies per country')

layoffs.country.value_counts()
```

```python
print(f'Total number of industries affected by layoffs: {len(layoffs.industry.unique())}')
print()
print('Number per industry:')
layoffs.industry.value_counts()
```

```python
layoffs.date.unique()
```

```python
len(layoffs.date.unique())

layoffs['date'] = pd.to_datetime(layoffs['date'])

print(f'Earliest date in the dataset {layoffs.date.min()}')
print(f'Latest date in the dataset {layoffs.date.max()}')
```

```python
import datetime as dt

layoffs['year'] = layoffs.date.dt.year
layoffs['month'] = layoffs.date.dt.month
layoffs['day'] = layoffs.date.dt.day

layoffs
```

```python
print(f'Total number of employees laid off are {layoffs.total_laid_off.sum()}')
```

```python
twenty_20 = layoffs.loc[layoffs['year'] == 2020]
twenty_21 = layoffs.loc[layoffs['year'] == 2021]
twenty_22 = layoffs.loc[layoffs['year'] == 2022]

print(f'Total number of layoffs in 2020: {twenty_20.total_laid_off.sum()}')
print()
print(f'Total number of layoffs in 2021: {twenty_21.total_laid_off.sum()}')
print()
print(f'Total number of layoffs in 2022: {twenty_22.total_laid_off.sum()}')
```

```python
country_comp = layoffs.groupby('country')['funds_raised', 'total_laid_off'].sum()
country_comp
```

```python
industry_comp = layoffs.groupby('industry')['funds_raised', 'total_laid_off'].sum()
industry_comp

plt.figure()
plt.scatter(industry_comp['total_laid_off'],industry_comp['funds_raised'])
plt.xlabel('Laid Off Employees')
plt.ylabel('Funds Raised')
plt.show()
```

```python
country_comp = layoffs.groupby('country')['funds_raised', 'total_laid_off'].sum()
country_comp

plt.figure()
plt.scatter(country_comp['total_laid_off'],country_comp['funds_raised'])
plt.xlabel('Laid Off Employees')
plt.ylabel('Funds Raised')
plt.show()
```
